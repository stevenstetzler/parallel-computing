```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
from matplotlib import rcParams
```

```python
rcParams['figure.figsize'] = [18, 12]
rcParams['font.size'] = 22
rcParams['xtick.major.size'] =  8
rcParams['xtick.minor.size'] =  4
rcParams['xtick.major.width'] =  2
rcParams['xtick.minor.width'] =  2
rcParams['ytick.major.size'] =  8
rcParams['ytick.minor.size'] =  4
rcParams['ytick.major.width'] =  2
rcParams['ytick.minor.width'] =  2
```

```python
threads = np.array([64, 32, 16, 8, 4, 2, 1])
threads_2 = np.array([64, 16, 4, 1])
type_0 = np.array([337, 434, 613, 866, 926, 1487, 2634])
type_1 = np.array([1150, 1176, 1034, 1199, 1129, 1483, 2634])
type_2 = np.array([732, 714, 970, 2634])
```

```python
plt.plot(threads, type_0, linestyle='none', marker='o', markersize=10, color='k', label="Assign Rows")
plt.plot(threads, type_1, linestyle='none', marker='^', markersize=10, color='grey', label="Assign Columns")
plt.plot(threads_2, type_2, linestyle='none', marker='s', markersize=10, color='blue', label="Assign Squares")

plt.xlabel("Number of Threads")
plt.ylabel("Run Time (s)")
plt.legend()

plt.show()
```

```python
plt.loglog(threads, type_0, linestyle='none', marker='o', markersize=10, color='k', label="Assign Rows")
plt.loglog(threads, type_1, linestyle='none', marker='^', markersize=10, color='grey', label="Assign Columns")
plt.loglog(threads_2, type_2, linestyle='none', marker='s', markersize=10, color='blue', label="Assign Squares")

plt.xlabel("Number of Threads")
plt.ylabel("Run Time (s)")
plt.legend()

plt.show()
```

```python
result = linregress(np.log10(threads), np.log10(type_0))
type_0_slope = result.slope
type_0_inter = result.intercept
print(type_0_slope, type_0_inter)
```

```python
x_line = np.linspace(1, 64, 1000)
y_line = np.power(10, type_0_slope * np.log10(x_line) + type_0_inter)

plt.loglog(threads, type_0, linestyle='none', marker='o', markersize=10, color='k', label="Assign Rows")
plt.loglog(x_line, y_line, linestyle='--', linewidth=3, color='k', label='Rows Fit')

plt.loglog(threads, type_1, linestyle='none', marker='^', markersize=10, color='grey', label="Assign Columns")
plt.loglog(threads_2, type_2, linestyle='none', marker='s', markersize=10, color='blue', label="Assign Squares")

plt.xlabel("Number of Threads")
plt.ylabel("Run Time (s)")
plt.legend()

plt.show()
```
